```python
# Weather dataset creation using numpy arrays
import numpy as np
import pandas as pd

cities = np.array([
    "Chennai","Hyderabad","Bengaluru","Mumbai","Delhi",
    "Chennai","Hyderabad","Bengaluru","Mumbai","Delhi",
    "Chennai","Hyderabad","Bengaluru","Mumbai","Delhi",
    "Chennai","Hyderabad","Bengaluru","Mumbai","Delhi",
    "Chennai","Hyderabad","Bengaluru","Mumbai","Delhi"
])

months = np.array([
    "Jan","Jan","Jan","Jan","Jan",
    "Feb","Feb","Feb","Feb","Feb",
    "Mar","Mar","Mar","Mar","Mar",
    "Apr","Apr","Apr","Apr","Apr",
    "May","May","May","May","May"
])

temperature = np.array([
    28,26,24,29,18,
    30,29,25,31,22,
    33,35,28,34,27,
    36,38,30,35,32,
    39,41,32,37,35
])

humidity = np.array([
    65,60,68,72,55,
    70,62,75,78,58,
    75,68,80,82,62,
    78,72,85,88,65,
    82,76,90,92,70
])

rain = np.array([
    15,10,30,35,5,
    20,18,40,45,12,
    45,25,65,70,35,
    30,28,75,80,40,
    85,35,90,95,50
])

weather = np.array([
    "Sunny","Sunny","Cloudy","Cloudy","Sunny",
    "Sunny","Sunny","Cloudy","Cloudy","Sunny",
    "Cloudy","Sunny","Rainy","Rainy","Cloudy",
    "Sunny","Sunny","Rainy","Rainy","Cloudy",
    "Rainy","Sunny","Rainy","Rainy","Cloudy"
])

df = pd.DataFrame({
    "City": cities,
    "Month": months,
    "Temperature": temperature,
    "Humidity": humidity,
    "Chance_of_Rain": rain,
    "Weather": weather
})

print("weather data :")
print(df)
```

```python
# the Top 3 Hottest City-Month Records
top3_idx = np.argsort(temperature)[-3:][::-1]

for idx in top3_idx:
    print(cities[idx], months[idx], temperature[idx])
```

```python
# to Determine Which City Has the Highest Average Humidity
unique_cities = np.unique(cities)

avg_humidity = []

for city in unique_cities:
    avg = np.mean(humidity[cities == city])
    avg_humidity.append(avg)

avg_humidity = np.array(avg_humidity)

highest = np.argmax(avg_humidity)

print("city:",unique_cities[highest])
print("Highest Average Humidity:",avg_humidity[highest])
```

```python
# 3. Monthly Average Temperature (Jan–May)
unique_months = np.unique(months)

for month in unique_months:
    avg_temp = np.mean(
        temperature[months == month]
    )

    print(month, round(avg_temp,2))
```

```python
# 4. City with Highest Weather Variability
variability = []

for city in unique_cities:
    std = np.std(
        temperature[cities == city]
    )

    variability.append(std)

variability = np.array(variability)

highest = np.argmax(variability)
print("city:",unique_cities[highest])
print("Highest Variability:",variability[highest])

```

```python
# 5. Humidity > 80 and Rain Chance > 70
mask = (
    (humidity > 80) &
    (rain > 70)
)

for city, month in zip(
    cities[mask],
    months[mask]
):
    print(city, month)
```

```python
 # 6. Rank Cities by Rain Intensity Score
 # formula: (Humidity × Chance of Rain) / 100
 score = (humidity * rain) / 100

city_scores = []

for city in unique_cities:

    avg_score = np.mean(
        score[cities == city]
    )

    city_scores.append(avg_score)

city_scores = np.array(city_scores)

rank = np.argsort(
    city_scores
)[::-1]

for i in rank:
    print(
        unique_cities[i],
        round(city_scores[i],2)
    )
```

```python
# 7. Cities Whose Temperature Is Above Their Own Average
for city in unique_cities:

    city_avg = np.mean(
        temperature[cities == city]
    )

    mask = (
        (cities == city) &
        (temperature > city_avg)
    )

    print("\n", city)

    for month, temp in zip(
        months[mask],
        temperature[mask]
    ):
        print(month, temp)
```

```python
#8. Coastal vs Inland Cities
coastal = np.isin(
    cities,
    ["Chennai","Mumbai"]
)

inland = np.isin(
    cities,
    ["Delhi","Hyderabad","Bengaluru"]
)
print("chennai:")
print(
    "Coastal Avg Temp:",
    np.mean(
        temperature[coastal]
    )
)
print(
    "Coastal Avg Humidity:",
    np.mean(
        humidity[coastal]
    )
)

print("mumbai")
print(
    "Inland Avg Temp:",
    np.mean(
        temperature[inland]
    )
)


print(
    "Inland Avg Humidity:",
    np.mean(
        humidity[inland]
    )
)
```

```python
# 9. Sunny Weather with Rain Chance > 30
condition = (
    (weather == "Sunny") &
    (rain > 30)
)

for city, month in zip(
    cities[condition],
    months[condition]
):
    print(city, month)
```

```python
# 10. Relationship Between Temperature and Humidity
correlation = np.corrcoef(
    temperature,
    humidity
)

print(
    "Correlation:",
    correlation[0,1]
)
# Interpretation
corr = correlation[0,1]

if corr > 0:
    print(
        "Temperature and humidity increase together."
    )
else:
    print(
        "Temperature and humidity move in opposite directions."
    )
```
